## Count Seconds in a Decade
Write a Python program to calculate the number of seconds in 10 years. Assume that we have 2 leap years in this period. Assign the result into a variable named `seconds_in_10_years`.

```python
years = 10
leap_years = 2
regular_years = years - leap_years

days_in_regular_year = 365
days_in_leap_year = 366

total_days = (regular_years * days_in_regular_year) + (leap_years * days_in_leap_year)

seconds_in_a_day = 24 * 60 * 60

seconds_in_10_years = total_days * seconds_in_a_day

print(f"Number of regular years: {regular_years}")
print(f"Number of leap years: {leap_years}")
print(f"Total days in 10 years: {total_days}")
print(f"Total seconds in 10 years: {seconds_in_10_years}")
```

## Compute Euclidean Distance
Write a Python program to compute the Euclidean distance between two given points $ (x_1, y_1) $ and $ (x_2, y_2) $, which is $ \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2} $.

```python
x1 = int(input('input x1: '))
y1 = int(input('input y1: '))
x2 = int(input('input x2: '))
y2 = int(input('input y2: '))

euclidean_distance =((x1 - x2)**2 + (y1 - y2)**2)**0.5

print(f"Euclidean distance between ({x1}, {y1}) and ({x2}, {y2}) is {euclidean_distance:}")
```

## Print User's Name
Write a Python program which accepts the user's first and last name and prints them in reverse order with a space between them. For example, if the user enters `John Doe`, the program should print `Doe John`. Note that you can use the `input()` function to take user's inputs.

```python
first_name = input("Enter first name: ")
last_name = input("Enter last name: ")

print(f"Original: {first_name} {last_name}")

# Reverse using a temporary variable
temp = first_name
first_name = last_name
last_name = temp

print(f"Reverse: {first_name} {last_name}")
```

## Read a Text File
Write a Python program that opens a text file named `my_file.txt` and writes `Hello world!` in that file.

```python
# Open a text file named 'my_file.txt' and write to it
file_name = 'my_file.txt'

# Open the file in write mode ('w')
# If the file doesn't exist, it will be created
# If it exists, it will be overwritten
with open(file_name, 'w') as file:
    file.write('Hello world!')

print(f"Successfully wrote 'Hello world!' to {file_name}")
```
